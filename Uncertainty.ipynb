```python
# _/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/
#  STEP 4 : What is Probability Distribution Function ?
#       
#         Just few parameters to predict thousands of value scatters !
#
# _/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/_/
#
#  Related Fields : [Mathematics]
#
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x_mean = 2.0
x_sd   = 1.0
xLim   = [-2,6]
nsamp  = 100

x1     = x_mean-2*x_sd
x2     = x_mean+2*x_sd

x      = np.arange( x1,  x2, 0.001 ) 
x_all  = np.arange( -10, 10, 0.001 ) 
y      = norm.pdf (     x, x_mean, x_sd )
y_all  = norm.pdf ( x_all, x_mean, x_sd )

xx1    = np.full(100,x1)
xx2    = np.full(100,x2)
yy1    = np.linspace(0, 0.055, 100)
yy2    = yy1

plt.figure(figsize=(14,5))
ax1   = plt.subplot(1,2,1)
ax1.plot         (yy1,xx1,color="b")
ax1.plot         (yy2,xx2,color="b")
ax1.plot         (y_all,x_all,color="b")
ax1.fill_between ( y, x, x_mean, color="lightblue" )
ax1.fill_between ( yy1, xx1, xx2, facecolor='lightblue',alpha=1.0, where=yy1<0.057 )

ax1.set_ylim  (xLim)
ax1.set_xlim  ([0.5,0])
ax1.set_xlabel('Probability Distribution Function PDF(X)')

ax2   = plt.subplot(1,2,2)
s     = np.random.normal(x_mean, x_sd, nsamp)
s     = np.random.normal(x_mean, x_sd, nsamp)

s_mean_list = [s[0]]
for i in range(1,len(s)):
    values = s[0:i]
    s_mean = np.average(values)
    s_mean_list.append(s_mean)

index = np.arange(0,nsamp,1)
ax2.scatter(index,s,color='b')
ax2.plot(index,s_mean_list,color='r')
ax2.set_xlabel("Test / Runs")
ax2.set_ylabel("X")
ax2.set_ylim(xLim)
ax1.grid(b=True, which='major', color='#666666', linestyle='-')
ax2.grid(b=True, which='major', color='#666666', linestyle='-')
plt.show()
```
